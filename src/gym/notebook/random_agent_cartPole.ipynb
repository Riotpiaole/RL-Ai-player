## Random Search For solutions

1. it randomly generate weights until it reaches some boundaries. ie: highest reward
2. The implementation goes over each eposides with a different weights hence `parameters`
3. algorithm goes 
```python
for eposide in nums_eposides:
    nW = generateParameters()
    prediction= nW*observation
    if prediction > 0:
        env.step(1)
    else:
        env.step(0)
```

- the goal is searching for solution with randomly generated w
- $\begin{aligned}
        y = W\times X \text{ such that W is randomly generated}
\end{aligned}$
